```python
#dependencies

import pretty_midi
import os
import numpy as np
import tensorflow as tf
import tensorflow.keras as keras
from keras.utils import to_categorical
```

```python
# Set directory path where MIDI files are located
midi_dir = "C:/Users/simas/Downloads/Midis by Alawais/midi dataset/"

# Create list of MIDI file paths
files = []
for file in os.listdir(midi_dir):
    if file.endswith(".mid"):
        files.append(os.path.join(midi_dir, file))
```

```python
print(files)
```

```python
# Load MIDI files into memory
midi_data = []
for file in files:
    midi_data.append(pretty_midi.PrettyMIDI(file))
```

```python
print(midi_data)
```

```python
# Extract MIDI messages from MIDI files
messages = []
for midi in midi_data:
    for instrument in midi.instruments:
        for note in instrument.notes:
            message = (note.pitch, note.velocity, note.start)
            messages.append(message)

messages = np.array(messages)
```

```python
print(messages)
```

```python
sequence_length = 32

sequences = []
for i in range(len(messages) - sequence_length):
    sequence = messages[i:i + sequence_length]
    sequences.append(sequence)
    
sequences = np.array(sequences)
```

```python
print(sequences)
```

```python
encoded_sequences = to_categorical(sequences)
```

```python
# Save the encoded sequences to a .npy file
np.save("encoded_sequences.npy", encoded_sequences)
```
